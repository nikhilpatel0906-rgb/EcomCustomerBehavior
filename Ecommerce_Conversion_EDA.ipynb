Online E-Commerce Customer Conversion Analysis

```python
# import libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

sns.set_style("whitegrid")
```

```python
# data load
df = pd.read_csv("online_shoppers_intention.csv")
df.head()
```

```python
df.shape
# df.info()
# df.describe()
```

```python
# checking missing value

df.isnull().sum()     #its already clean data from source
```

```python
df['Revenue'] = df['Revenue'].astype(int)
df['Weekend'] = df['Weekend'].astype(int)
# convert True False to 0,1
```

```python
df['Month'] = df['Month'].astype('category')
```

```python
# Overall Conversion Rate

conversion_rate = df['Revenue'].mean() * 100
print("Conversion Rate:", round(conversion_rate,2), "%")
```

```python
# Visitor Type Analysis
plt.figure(figsize=(6,4))
sns.countplot(x='VisitorType', hue='Revenue', data=df)
plt.title("Visitor Type vs Purchase")
plt.show()
```

```python
# WeekEnd Effect
plt.figure(figsize=(6,4))
sns.countplot(x='Weekend', hue='Revenue', data=df)
plt.title("Weekend vs Purchase")
plt.show()
```

```python
# Bounce Rate

plt.figure(figsize=(7,4))
sns.boxplot(x='Revenue', y='BounceRates', data=df)
plt.title("Bounce Rate vs Purchase")
plt.show()
```

```python
# Exit Rate

plt.figure(figsize=(7,4))
sns.boxplot(x='Revenue', y='ExitRates', data=df)
plt.title("Exit Rate vs Purchase")
plt.show()
```

```python
# Engagement Analysis

plt.figure(figsize=(7,5))
sns.scatterplot(x='ProductRelated_Duration', y='PageValues',
                hue='Revenue', data=df)
plt.title("Product Engagement vs Purchase")
plt.show()
```

```python
# Monthly Trend

plt.figure(figsize=(10,4))
sns.countplot(x='Month', hue='Revenue', data=df,
              order=df['Month'].value_counts().index)
plt.xticks(rotation=45)
plt.title("Monthly Purchase Trend")
plt.show()
```

```python
# Correlation Heatmap
plt.figure(figsize=(12,8))
sns.heatmap(df.corr(numeric_only=True), cmap='coolwarm', annot=False)
plt.title("Feature Correlation Heatmap")
plt.show()
```
